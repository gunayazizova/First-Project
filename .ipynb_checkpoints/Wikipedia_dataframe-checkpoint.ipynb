```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
    url = 'https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_Human_Development_Index_score' 
    response = requests.get(url)
    tables = pd.read_html(url)
    soup = BeautifulSoup(response.text, 'html.parser')
```

```python
exclude_list = ['District of Columbia', 'Guam', 'Guam[2]', 'United States (average)', 'U.S. Virgin Islands[2]', 'Northern Mariana Islands[2]', 'American Samoa[2]', 'U.S. Virgin Islands', 'Puerto Rico', 'Northern Mariana Islands', 'American Samoa']

# Clean the 'State' column
#df_cleaned = df[~df['State, federal district, or territory'].isin(exclude_list)]
df_cleaned = df.drop(df[df['State, federal district,  or territory'].isin(exclude_list)].index)

# Display the cleaned DataFrame
df_cleaned
```
